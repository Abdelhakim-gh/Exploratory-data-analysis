# EDA: le taux de désabonnement des banques

## Apreçu sur problem

*Définition : Qu'est-ce que l'attrition ?*
- Pour les besoins de ce tutoriel, nous parlerons simplement d'attrition lorsqu'un **client annule un contrat de service avec la banque**.

*Pourquoi l'attrition est-elle si importante ?*
- Ce n'est qu'en comprenant l'objectif final que vous pourrez construire un modèle utile. 
- Il s'agit d'identifier les clients qui risquent de se désabonner.
- d'identifier les clients à risque d'attrition.

| Total_Ct_Chng_Q4_Q1     | dbl             | Similaire au précédent mais en termes de nombre de transactions                                                                                                            |
| Avg_Utilization_Ratio    | dbl             | Représente la proportion du crédit disponible que le client a dépensée                                                                                                      |


**Detaile :**


1. **Total_Relationship_Count** : Ce chiffre indique le nombre de produits bancaires différents (comptes d'épargne, cartes de crédit, prêts) qu'un client possède auprès de la banque. Plus il y a de produits, plus la relation avec la banque est profonde.

2. **Total_Revolving_Balance** : Il s'agit du montant que vous devez sur votre carte de crédit et que vous n'avez pas encore remboursé. Si vous achetez quelque chose avec votre carte de crédit et que vous n'avez pas encore payé la facture, c'est votre solde renouvelable.

3. **Avg_Open_To_Buy** : Il s'agit du montant que vous pouvez dépenser sur votre carte de crédit avant d'atteindre votre limite de crédit. C'est un peu comme votre limite de dépenses.

4. **Total_Trans_Amt** : Ce montant nous indique combien d'argent vous avez dépensé ou transféré sur vos comptes au cours de l'année écoulée. Cela comprend les achats, les paiements et les autres activités financières.

4. **Total_Trans_Ct** : Nombre total de transactions count.

5. **Total_Amt_Chng_Q4_Q1**: Cette question indique si vos habitudes de dépenses ont changé entre la fin de l'année dernière et le début de cette année. Avez-vous dépensé plus ou moins ?

6. **Total_Ct_Chng_Q4_Q1**: Semblable à la précédente, mais au lieu de la somme d'argent dépensée, il s'agit du nombre de transactions que vous avez effectuées. Avez-vous acheté plus ou moins de choses ?

7. **Ratio d'utilisation moyenne** : Il s'agit de la proportion de la limite de votre carte de crédit que vous avez utilisée. Si votre limite de crédit est de 1 000$ et que vous avez dépensé 500 $, votre taux d'utilisation est de 50 %. Il aide les banques à comprendre quelle part de votre crédit disponible vous utilisez.

## Analyse univarie & bivaraie de données Catégoriale

### Analyse univarie [X]

- *Frequence / count* pour caractere (X)

```python
categoricals = ['Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']

def pltCountplot(data, cats):
    num_cols = 3
    num_rows = (len(cats) + num_cols - 1) // num_cols  # Calculate the number of rows needed
    fig, axis = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))  # Adjust figsize based on the number of rows
    index = 0
    for i in range(num_rows):
        for j in range(num_cols):
            if index < len(cats):
                ax = sns.countplot(data=data, x=cats[index], ax=axis[i][j], palette='Set2', hue=cats[index], legend=False)  # Specify color palette here
                if cats[index] in ['Education_Level', 'Income_Category']:
                    for item in ax.get_xticklabels():
                        item.set_rotation(15)
                
                for p in ax.patches:
                    height = p.get_height()
                    ax.text(
                            p.get_x()+p.get_width()/2.,
                            height + 3,
                            '{:1.2f}%'.format(height/len(data)*100),
                            ha="center"
                        ) 
                index += 1
pltCountplot(data_set, categoricals)
```

*Remarque :*
- La plupart des clients sont principalement Existants, représentant *83,93%* -> 8 499 clients.
- La plupart des clients :
    - Ont obtenu leur diplôme ou ont terminé le lycée.
    - Sont mariés ou célibataires.
    - Ont un revenu inférieur à 40k ou compris entre 40k et 60k.
    - Possèdent une carte bleue.
